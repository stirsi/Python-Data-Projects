```python
#create new column names

new_cols_dict = {
    'timestamp' : 'time', 
    'cnt' : 'count', 
    't1':'temp_real_C',
    't2': 'temp_feels_like_C', 
    'hum' : 'humidity_percent', 
    'wind_speed' : 'wind_speed_kph', 
    'weather_code': 'weather', 
    'is_holiday' : 'is_holiday', 
    'is_weekend': 'is_weekend', 
    'season' : 'season'
}

#rename columns
bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
bikes
```

```python
#changing humidity values to percentage
bikes.humidity_percent = bikes.humidity_percent / 100 
```

```python
#create a season dictionary to map the integers 0-3 to the actual written values

season_dict = {
    '0.0': 'spring',
    '1.0': 'summer', 
    '2.0' : 'autumn', 
    '3.0' : 'winter'
}

#create a weather dictionary to map the integers to actual written values
weather_dict = {
    '1.0': 'Clear', 
    '2.0' : 'Scattered clouds',
    '3.0' : 'Broken clouds', 
    '4.0' : 'Cloudy',
    '7.0' : 'Rain',
    '10.0' : 'Rain with thunderstorm',
    '26.0': 'Snowfall'
}
#changing season column data type to string
bikes.season = bikes.season.astype('str')
#mapping the values 0-3 to the actual written seasons
bikes.season = bikes.season.map(season_dict)

#changing the weather column data type to string
bikes.weather = bikes.weather.astype('str')
#mapping the codes to actual written weather values
bikes.weather = bikes.weather.map(weather_dict) 
```
